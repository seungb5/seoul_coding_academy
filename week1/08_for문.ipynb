#### for문을 이용한 구구단
- 1단부터 만들어보자.
- 2단도 만들어보자.
- 일반화 해보자.

```python
l1 = list(range(1, 10))
l1
```

```python
l2 = [] # for문을 써보자.
for i in range(1,10):
    l2.append(i)
    print(l2)
l2 # 들어가는 과정 참고.
```

```python
l3 = [] # 구구단 만들기
for j in range(1,10):
    for i in range(1,10):
        l3.append(i*j)
        print(l3)
```

```python
list99 = [] # 10*10 행렬로 구구단 만들기
for j in range(1,10):
    l = []
    for i in range(1,10):
        l.append(i*j)
        print(l)
    list99.append(l)
    print(list99)
```

```python
# Result
l = [3, 6, 7, 9, 2, 8, 5, 1] # 정렬해보자
n = len(l)
sortedl = []
```

```python
for j in range(n):
    m = l[0]
    for i in range(len(l)):
        m = min(m, l[i])
    sortedl.append(m)
    l.remove(m)
    print(l, sortedl)
```

```python
l = [3, 6, 7, 9, 2, 8, 5, 1]
n = len(l)
sortedl = []
```

```python
for j in range(n):
    m = 0
    for i in range(len(l)):
        if l[i] < l[m]: m = i
    sortedl.append(l.pop(m))
    print(l, sortedl)
```
