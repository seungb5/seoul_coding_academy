### X,y train

```python
# 1. drop 
train_drop = train_data.drop(labels=['PassengerId', 'Name', 'Cabin', 'Ticket', 'Embarked', 'Sex'], axis = 1)
train_drop.head()
```

### X, y test

```python
# drop
X_test_drop = X_test_data.drop(labels=['PassengerId', 'Name', 'Cabin', 'Ticket', 'Embarked', 'Sex'], axis = 1)
```

```python
# 결측치 대체
DataFrameSummary(X_test_drop).columns_stats
```

- 결측된 열이 2 가지(Age, Fare) 존재하며, 이를 평균값으로 대체하자

```python
X_test_drop.Age.fillna(value=Age_mean, inplace=True)
X_test_drop.Fare.fillna(value=train_drop.Fare.mean(), inplace=True)
```

```python
X_test = X_test_drop
y_test = y_test_data
```

```python
# KNN 적용하기
```

```python
X_test
```

```python
training_accuracy = []
test_accuracy = []
# 1 에서 10 까지 n_neighbors 를 적용
neighbors_settings = range(1, 50)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="train accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```
