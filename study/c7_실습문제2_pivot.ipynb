```python
def mean(df):
    return pd.Series(np.array([sum(df.iloc[i])/3 for i in range(20)]), index = range(1, 21), name = 'mean')
```

```python
# 실습 답안
col = ['math', 'eng', 'sci']
ClassA = pd.DataFrame(data = np.array([random.randint(30,100) for i in range(30)]).reshape(10, 3)
                        , index = range(1, 11), columns = col)

ClassB = pd.DataFrame(data = np.array([random.randint(30,100) for i in range(30)]).reshape(10, 3)
                        , index = range(1, 11), columns = col)

df = pd.concat([ClassA, ClassB], keys = ['A', 'B'], names= ['class', 'student_id'])
df.head(5)
```

```python
# 할 필요 없지만 pivot_table 해보기.
df["mean"] = np.mean(df, axis=1) # df에 col 추가하는 방법. 너무 깔끔하당.
df_ex = df.reset_index()
df_ex[:5]
```

```python
df_ex.pivot_table(values="mean", index=["class", "student_id"])[:5]
```

```python
# 실습 답안
col = ['math', 'eng', 'sci']
ClassA = pd.DataFrame(data = np.array([random.randint(30,100) for i in range(30)]).reshape(10, 3)
                        , index = range(1, 11), columns = col)

ClassB = pd.DataFrame(data = np.array([random.randint(30,100) for i in range(30)]).reshape(10, 3)
                        , index = range(1, 11), columns = col)

df = pd.concat([ClassA, ClassB], keys = ['A', 'B'], names= ['class', 'student_id'])
df.head(5)
```

```python
# 내 방법 : groupby도 사용하지 않았고, 쓸데 없이 reset_index, join, pivot_table을 사용하게되는 비효율.
mean = pd.DataFrame(df.mean(axis = 1), columns=['mean'])
mean[:5]
```

```python
df2 = df.join(mean, how = 'outer')
df3 = df2.sort_values(by = 'mean', ascending=False)
df3.head(5)
```

```python
df4 = df3.reset_index()
df4
```

```python
df5 = df4.pivot_table(index=['class', 'student_id'],  values='mean')
df5
```

```python
a = df5.reset_index().sort_values(by = ['class', 'mean'], ascending= [True, False])
a[5:15]
```
