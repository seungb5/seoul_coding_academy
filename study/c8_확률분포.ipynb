## 확률 분포
- binomial(n, p[, size]) : p 확률 이항분포에서 n번 시행할 때, positive 빈도를 size 만큼의 샘플로 추출.
- multinomial(n, pvals, size = None) : size x p == 횟수 x 범주
- multivariate_normal(mean, cov[, size, check_valid, tol]) : shape의 다변량 정규분포 샘플.
- 공분산과 상관계수의 차이 : 평균과 분산을 고정하였는지 여부.

### 정규분포
- 자연계에서 일상적으로 나타나는 분포
- 의도가 없음. 만약 의도가 있다면 skewed된 분포가 나올 것.
- 백색 잡음. white noise
- 표준편차 : 평균으로 부터 얼마나 벗어나 있는가

### 공분산과 상관계수의 차이 : 평균과 분산을 고정하였는지 여부.

```python
import numpy as np
```

```python
np.random.binomial(100, 1/2., size = 4)
```

```python
np.random.multinomial(100, [1/7.]*5 + [2 /7.], size = 4)
```

- 아래는 2차원 다변량 정규분포에서 평균과 공분산의 조건아래 분포의 샘플.

```python
import matplotlib.pyplot as plt
mean = [0, 0]
cov = [[5, -5], [-5, 20]]

x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
