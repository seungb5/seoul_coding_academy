#### 문제 정의
- 반별 과목별 평균을 구하고, 반별 평균보다 높은 학생 id를 출력하라.

```python
# class와 id에 따른 pivoting
df = data.pivot_table(index=['class', 'id'])
df.head()
```

```python
# 학생의 과목별 평균 구하기.
df["mean"] = np.mean(df, axis = 1)
df
```

```python
# 반 별 grouping
df2 = df["mean"].groupby(level = 0, group_keys=False) # group_keys....?
```

```python
for group in df2:
    print(group)
```

```python
df2.mean()
```

```python
result = pd.DataFrame(df2.apply(lambda x: x[x > x.mean()])) # x[x > x.mean()] : fancy indexing
result
```
