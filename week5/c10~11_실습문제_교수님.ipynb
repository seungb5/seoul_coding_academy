### 1) DataFrame의 결합
6장에서 생성한 call_data와 put_data를 이용하여 두 개의 DataFrame을 결합하는 방법을 적용해본다.

```python
url = 'https://finance.yahoo.com/quote/AAPL/options?ltr=1'

from lxml.html import parse
from urllib.request import urlopen

parsed = parse(urlopen(url))

doc = parsed.getroot()
tables = doc.findall('.//table')
call = tables[0]; put = tables[1]
```

```python
def _unpack(row, tag='td'):
    tds = row.findall('.//%s' % tag)
    rec = [td.text_content() for td in tds]
    return rec
```

```python
from pandas.io.parsers import TextParser
```

```python
def parse_data_from_table(tb):
    rows = tb.findall('.//tr')
    head = _unpack(rows[0], tag='th')
    data = [_unpack(row) for row in rows[1:]]
    parsed = TextParser(data, names=head)
    return parsed.get_chunk()
```

```python
call_data = parse_data_from_table(call)
put_data = parse_data_from_table(put)
```

#### 5) iddf의 반별 과목별 id를 ranks의 반별 과목별 순위에 맞춰 정렬.
- set_index보다는 set_values의 by 옵션을 사용하는 것이 적절해보인다.
- 이를 위해 먼저 반 별로 하나의 과목에 대해 정렬하는 방법을 알아보자.
- set_values를 사용하기 위해 iddf와 ranks를 반별 과목별로 join하자.

join 메소드는 DataFrame의 메소드이므로, DataFrame을 유지하도록 math 과목을 선택한다.

```python
merged = ranks[['math']].join(iddf[['math']], lsuffix='_r')
merged
```

먼저 특정 반 별로 math 과목에 대해 iddf를 ranks로 정렬해보자.

```python
megroups = merged.groupby('class')
i = 0
for cls, megroup in megroups:
    if i == 3:
        break
    i += 1
meg_math4 = megroup
meg_math4
```

```python
'%s_r' % 'math'
```

```python
def sort_by_rank(mgroup, subj):
    sort_col = '%s_r' % subj
    tmp = mgroup.sort_values(by=sort_col)
    tmp[subj].index = [cls]*tmp.shape[0]
    return tmp[subj]
```

```python
sort_by_rank(meg_math4, 'math')
```

이제 이를 megroups에 있는 모든 group에 적용해보자. 이를 위해 아래와 같은 연산의 결과가 for 문에 의해 어떻게 생성되는 지 확인해보자.

```python
for subj in subjs:
    merged = ranks[[subj]].join(iddf[[subj]], lsuffix='_r')
    megroups = merged.groupby('class')
    for cls, group in megroups:
        subj_tmp = sort_by_rank(group, subj)
        print(subj_tmp)
```

먼저 각 반에 대해 우선 과목별로 id를 rank 순으로 정렬한다. 따라서:
1. megroups에 대한 순회 결과는 pd.concat으로 수직결합한다.
2. subjs에 대한 순회 결과는 pd.concat으로 수평결합한다.

```python
lastlist = []
for subj in subjs:
    merged = ranks[[subj]].join(iddf[[subj]], lsuffix='_r')
    megroups = merged.groupby('class')
    intermediate = []
    for cls, group in megroups:
        subj_tmp = sort_by_rank(group, subj)
        intermediate.append(subj_tmp)
    class_tmp = pd.concat(intermediate)
    lastlist.append(class_tmp)
last = pd.concat(lastlist, axis=1)
last
```

```python
last.dropna(how='all')
```

## 9장. 실습
### 9.1 group 연산에 대한 다양한 방법
1. tips 데이터에서 grouped를 생성한다.
 - group key로 ‘sex’와 ‘day’를 지정
2. 연속형 변수들(bill, tip 등)은 mean을 범주형 변수는 count를 적용한다.
 - agg을 통해 구현한다.
 - apply를 통해 구현한다.
     - 필요시 concat를 사용한다.
3. pivot_table로 구현한다.
 - 필요시 concat를 사용한다.
 - pivot_table 사용 시, 중간합계를 계산한다.
 
#### 1) agg로 구현하면

```python
tips
```

```python
group_key = ['sex', 'day']
```

```python
grouped = tips.groupby(group_key)
```

```python
tips.columns
```

```python
tips.dtypes
```

```python
str_cols = []
num_cols = []
new_num_cols = []
new_str_cols = []
for col in tips.columns:
    if col in group_key:
        continue
    if tips[col].dtype in [np.float64, np.int64]:
        num_cols.append(col)
        new_num_cols.append(col + "_mean")
    else:
        str_cols.append(col)
        new_str_cols.append(col + "_count")
```

```python
num_cols
```

```python
str_cols
```

```python
mean_new_dict = [(col, f) for col, f in zip(new_num_cols, ['mean']*len(new_num_cols))]
mean_new_dict
```

```python
mean_dict = {col:f for col, f in zip(num_cols, ['mean']*len(num_cols))}
mean_dict
```

```python
count_dict = {col:f for col, f in zip(str_cols, ['count']*len(str_cols))}
count_dict
```

```python
agg_functs = dict([(k, v) for k, v in mean_dict.items()] + 
                  [(k, v) for k, v in count_dict.items()])
agg_functs
```

```python
agg_functs = dict(mean_dict, **count_dict)
agg_functs
```

```python
grouped.agg(agg_functs)
```

#### 2) apply로 구현하면

```python
tips.columns
```

```python
def mean_count(group):
    d = {}
    for col in num_cols:
        d[col] = group[col].mean()
    for col in str_cols:
        d[col] = group[col].count()
    return d
```

```python
mean_count(tips)
```

```python
apply_ops = grouped.apply(mean_count)
apply_ops
```

```python
final_app = DataFrame(list(apply_ops.values))
final_app.index = apply_ops.index
final_app
```

#### 3) pivot_table 활용

```python
tips
```

```python
mean_df = tips.pivot_table(values=num_cols, index=group_key, 
                           aggfunc=np.mean, margins=True)
count_df = tips.pivot_table(values=str_cols, index=group_key, 
                            aggfunc=len, margins=True)
```

```python
pd.concat([mean_df, count_df], axis=1)
```
